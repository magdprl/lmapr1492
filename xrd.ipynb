```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.ext.matproj import MPRester

with MPRester("CN7BQQV6jKvoIlVgU18a") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-7173")

#Création d'un objet XRDCalculator
xrd = XRDCalculator(1.54060)
pat = xrd.get_pattern(structure)
print(pat)
```

```python
#premier pic

print("Angle de diffraction = ", pat.x[0])
print("Intensité = ", pat.y[0])
print("Indices hkl associés: ", pat.hkls[0])
```

```python
#deuxième pic

print("Angle de diffraction = ", pat.x[1])
print("Intensité = ", pat.y[1])
print("Indices hkl associés: ", pat.hkls[1])
```

```python
#troisième pic

print("Angle de diffraction = ", pat.x[2])
print("Intensité = ", pat.y[2])
print("Indices hkl associés: ", pat.hkls[2])
```
